## Tester /detect_panels_texts

```python
import requests

# URL du serveur Flask
url = "http://127.0.0.1:5000/detect_panels_texts"

# Chemin de l'image à tester
image_path = "test_files/test_image.jpg"

# Envoyer une requête POST avec l'image
with open(image_path, "rb") as image_file:
    response = requests.post(url, files={"file": image_file})

# Afficher la réponse
if response.status_code == 200:
    print("Détections YOLO :", response.json())
else:
    print("Erreur :", response.json())
```

## Tester /text_to_speech

```python
import requests
from IPython.display import Audio, display

# URL du serveur Flask
url = "http://127.0.0.1:5000/text_to_speech"

# Texte à convertir en audio
payload = {
    "text": "Bonjour, ceci est un test de synthèse vocale.",
    "speaker": "Damien Black",
    "language": "fr"
}

# Envoyer une requête POST avec le texte
response = requests.post(url, data=payload)

# Lire l'audio directement dans Jupyter Notebook
if response.status_code == 200:
    # Utiliser IPython pour lire le contenu de l'audio
    audio_data = response.content
    display(Audio(data=audio_data, autoplay=True))
else:
    print("Erreur :", response.json())
```
